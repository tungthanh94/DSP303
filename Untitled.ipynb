```python
import pandas as pd
import numpy as np
```

```python
#Storing the movie information into a pandas dataframe
movies_df = pd.read_csv('movies.csv')
#Storing the user information into a pandas dataframe
ratings_df = pd.read_csv('ratings.csv')
```

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))', expand= False)
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)', expand = False)
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '', regex= True)
movies_df['title'] = movies_df.title.apply(lambda x: x.strip())
movies_df.head()
```

```python
movies_df['genres'] = movies_df.genres.str.split('|')
```

```python
movie_w_genres = movies_df.copy()
```

```python
movie_w_genres.head()
```

```python
for index, val in movies_df.genres.iteritems():
    for genre in val:
        movie_w_genres.at[index, genre] = 1
```

```python
movie_w_genres.head()
```

```python
movie_w_genres.fillna(0, inplace = True)
```

```python
ratings_df.head()
```

```python
ratings_df.drop('timestamp', axis = 1, inplace=True)
```

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5},
            {'title':'Waiting to Exhale', 'rating':3.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
inputMovies = pd.merge(inputMovies, movies_df, on = 'title')
```

```python
inputMovies.drop(['genres', 'year'], axis = 1, inplace=True)
```

```python
inputMovies
```

```python
user_movies = movie_w_genres[movie_w_genres.movieId.isin(inputMovies.movieId)]
```

```python
user_movies
```

```python
#Resetting the index to avoid future issues
user_movies.reset_index(drop=True, inplace=True)
#Dropping unnecessary issues due to save memory and to avoid issues
userGenreTable = user_movies.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
userGenreTable
```

```python
inputMovies['rating'].shape
```

```python
userGenreTable.shape
```

```python
userGenreTable.T
```

```python
user_profile = userGenreTable.T.dot(inputMovies['rating'])
```

```python
user_profile
```

```python
#Now let's get the genres of every movie in our original dataframe
genreTable = movie_w_genres.set_index('movieId')
```

```python
genreTable
```

```python
#And drop the unnecessary information
genreTable = genreTable.drop('title', 1).drop('genres', 1).drop('year', 1)
genreTable.head()
```

```python
genreTable.shape
```

```python
user_profile
```

```python
recommend_df =  genreTable.dot(user_profile)
```

```python
top_recom = recommend_df.sort_values(ascending=False)
```
