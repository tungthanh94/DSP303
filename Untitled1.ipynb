```python
#Dataframe manipulation library
import pandas as pd
#Math functions, we'll only need the sqrt function so let's import only that
from math import sqrt
import numpy as np
```

```python
#Storing the movie information into a pandas dataframe
movies_df = pd.read_csv('movies.csv')
#Storing the user information into a pandas dataframe
ratings_df = pd.read_csv('ratings.csv')
```

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))', expand = False)
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)
#Removing the years from the 'title' column
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '', regex = True)
#Applying the strip function to get rid of any ending whitespace characters that may have appeared
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
```

```python
movies_df.head()
```

```python
movies_df.drop('genres', 1, inplace=True)
```

```python
ratings_df = ratings_df.drop('timestamp', 1)
```

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
inputMovies = pd.merge(inputMovies, movies_df, on = 'title')
inputMovies.drop('year', 1, inplace = True)
```

```python
inputMovies
```

```python
userSubset = ratings_df[ratings_df.movieId.isin(inputMovies.movieId)]
```

```python
userSubsetGroup = userSubset.groupby('userId')
```

```python
userSubsetGroup.get_group(1130)
```

```python
userSubsetGroup = sorted(userSubsetGroup, key = lambda x:len(x[1]), reverse = True)
```

```python
userSubsetGroup = userSubsetGroup[:100]
```

```python
userSubsetGroup
```

```python
inputMovies = inputMovies.sort_values(by = 'movieId')
```

```python
inputMovies
```

```python
from scipy.stats import pearsonr
```

```python
pearsonCorrelationDict = {}
```

```python
for name, group in userSubsetGroup:
    group  = group.sort_values(by = 'movieId')
    pearsonCorrelationDict[name] = pearsonr(inputMovies.rating, group.rating)[0]
```

```python
pearsonCorrelationDict
```

```python
pearsonDF = pd.DataFrame(data = pearsonCorrelationDict.values(), index=pearsonCorrelationDict.keys(),
            columns=['similarityIndex'])
```

```python
topUser = pearsonDF.sort_values(by = 'similarityIndex', ascending=False).head(50)
```

```python
topUser['userId'] = topUser.index
```

```python
topUser.reset_index(drop = True, inplace = True)
```

```python
topUsersRating = topUser.merge(ratings_df)
```

```python
topUsersRating
```

```python
topUsersRating['weightedRating'] = topUsersRating.similarityIndex*topUsersRating.rating
```

```python
tempTopUsersRating = topUsersRating.groupby('movieId').sum()[['similarityIndex','weightedRating']]
```

```python
tempTopUsersRating.columns = ['sum_similarityIndex','sum_weightedRating']
tempTopUsersRating.head()
```

```python
tempTopUsersRating['weighted average recommendation score'] = tempTopUsersRating['sum_weightedRating']/tempTopUsersRating['sum_similarityIndex']
```

```python
tempTopUsersRating
```

```python
tempTopUsersRating.sort_values(by = 'weighted average recommendation score', ascending=False, inplace=True)
```
